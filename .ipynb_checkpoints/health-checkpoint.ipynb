```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('com.samsung.shealth.sleep.202201142349.csv', skiprows = 1,index_col = False)
pd.set_option('display.max_columns', None)
```

```python
df.index = range(len(df))
```

```python
df['time_awake'] = df.sleep_duration- (df.efficiency / 100 * df.sleep_duration)
df['true_sleep_duration'] = df.sleep_duration * (df.efficiency/100)
```

```python
df['day'] = df['com.samsung.health.sleep.end_time'].str.split(' ').str[0]
```

```python
df.sleep_duration.plot()
```

```python
df[df.day == '2021-11-24']
```

```python
df
```

```python
ss_df = df.copy()
```

```python
ss_df = ss_df.groupby(by = 'day', as_index = False)['true_sleep_duration'].sum()
```

```python
ss_df
```

```python
import matplotlib.pyplot as plt
plt.scatter(ss_df.index, ss_df.true_sleep_duration)
```

```python
csv = 'com.samsung.shealth.step_daily_trend.202201142349.csv'
step_df = pd.read_csv(csv, skiprows = 1,index_col = False)
step_df
```

```python
step_df['day'] = step_df['create_time'].str.split(' ').str[0]
```

```python
step_df = step_df.groupby(by = 'day', as_index = False)['count'].sum()
```

```python
f_df = pd.merge(ss_df, step_df)
f_df['true_sleep_duration'] = f_df['true_sleep_duration']/60
```

```python
f_df
```

```python
bins = [i for i in range(0, int(f_df['count'].max()),int(f_df['count'].max()/12))]
f_df['binned'] = pd.cut(f_df['count'], bins, labels = [i for i in range(len(bins)-1)])
```

```python
f_df
```

```python
plt.scatter(f_df.index, f_df.true_sleep_duration, c = f_df['binned'], cmap = plt.cm.get_cmap('Blues'))
plt.colorbar(label = 'number of steps in 1000s')
plt.ylabel('hours of sleep')
plt.xlabel('day')
plt.show()
```
