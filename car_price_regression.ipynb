```python
import pandas as pd
import tensorflow as tf
from sklearn.preprocessing import OneHotEncoder
import numpy as np
```

```python
df = pd.read_csv('CarPrice_Assignment.csv')
```

```python
#df.symboling = df.symboling.apply(lambda x: f's{x}')
```

```python
one_hot = OneHotEncoder()
one_hot.fit(X = np.array(df.symboling).reshape(-1,1))
data = one_hot.transform(np.array(df.symboling).reshape(-1,1))
one_hot.categories_
dum_df = pd.DataFrame(data = data.toarray(), columns = one_hot.categories_)
```

```python
#df = df.join(dum_df)
```

```python
X = df.drop(columns = ['price', 'car_ID', 'CarName'])
y = df.price
```

```python
dummy_X = pd.get_dummies(X)
```

```python
#normal_X = (dummy_X-dummy_X.min())/(dummy_X.max()-dummy_X.min())
normal_X = dummy_X
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(normal_X, y, test_size = 0.2, random_state = 42)
```

```python
X_train.shape
```

```python
X_train
```

```python
tf.random.set_seed(42)
model = tf.keras.Sequential([
    tf.keras.layers.Normalization(axis = -1),
    tf.keras.layers.Dense(256, activation = 'relu'),
    tf.keras.layers.Dense(128, activation = 'relu'),
    tf.keras.layers.Dense(64, activation = 'relu'),
    tf.keras.layers.Dense(32, activation = 'relu'),
    tf.keras.layers.Dense(16, activation = 'relu'),
    tf.keras.layers.Dense(1)
    
])

model.compile(loss='mae',
              optimizer = tf.keras.optimizers.Adam(),
              metrics = ['mae']
             )

callback = tf.keras.callbacks.EarlyStopping(monitor='val_loss',
                              patience=10)

model.fit(X_train, y_train, validation_data = (X_test,y_test), epochs =1000, callbacks = callback)
```

```python
model.evaluate(X_test,y_test)
```
